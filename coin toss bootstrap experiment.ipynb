```python
import numpy as np
```

```python
total_tosses=30
experiment=np.random.randint(0,2,total_tosses)

print("Heads in the experiment:", experiment[experiment==1].shape[0])
```

### Coin Toss Experiment 1

```python
def coin_toss(times_to_repeat):
    head_count=np.empty((times_to_repeat,1))
    for times in range(times_to_repeat):
        experiment=np.random.randint(0,2,total_tosses)
        head_count[times]=experiment[experiment==1].shape[0]
    return head_count
```

```python
head_count=coin_toss(100000)
```

```python
head_count.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.distplot(head_count)
```

```python
head_count[head_count>=24]
```

```python
print("Mean:",np.mean(head_count))
print("Median:",np.median(head_count))
print("Standart deviation:",np.std(head_count))
print("No of times experiment returned 24 heads or more:", head_count[head_count>=24].shape[0])
print("% of times with 24 or more heads: ", head_count[head_count>=24].shape[0]/float(head_count.shape[0])*100)
```

### Coin Toss Experiment 2

```python
def coin_toss_experiment_2(times_to_repeat):

    head_count = np.empty([times_to_repeat,1], dtype=int)
    experiment = np.random.randint(0,2,[times_to_repeat,total_tosses])
    return experiment.sum(axis=1)
```

```python
head_count2=coin_toss_experiment_2(100000)
```

```python
sns.distplot(head_count2)
```

```python
print("Mean:",np.mean(head_count2))
print("Median:",np.median(head_count2))
print("Standart deviation:",np.std(head_count2))
print("No of times experiment returned 24 heads or more:", head_count2[head_count2>=24].shape[0])
print("% of times with 24 or more heads: ", head_count2[head_count2>=24].shape[0]/float(head_count2.shape[0])*100)
```
